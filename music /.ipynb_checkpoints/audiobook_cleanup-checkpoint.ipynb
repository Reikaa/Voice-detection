```python
# We'll need numpy for some mathematical operations
import numpy as np
import json

# Librosa for audio
import librosa

# matplotlib for displaying the output
import matplotlib.pyplot as plt
%matplotlib inline
# And seaborn to make it look nice
import seaborn
seaborn.set(style='ticks')

# and IPython.display for audio output
import IPython.display
```

```python
audio_path = './music/sound_input.wav'
```

```python
input1 = []
seconds_to_get = 1221
i = 0
while i < seconds_to_get:
    y, sr = librosa.load('sound_input.wav', duration=3.0, offset=i)
    i += 3
    input1.append(y)
    
print len(input1)

```

```python
# look at data
sampleM = input1[2]
sampleF = input1[207]
```

```python
# print('HAS_SAMPLERATE: ', librosa.core.audio._HAS_SAMPLERATE)
IPython.display.Audio(data=sampleM, rate=sr)
IPython.display.Audio(data=sampleF, rate=sr)
```

```python
M_harmonic, M_percussive = librosa.effects.hpss(sampleM) # 13s
F_harmonic, F_percussive = librosa.effects.hpss(sampleF) # 13s
M = librosa.feature.chroma_cqt(y=M_harmonic, sr=sr) # 5s
F = librosa.feature.chroma_cqt(y=F_harmonic, sr=sr) # 5s

# Make a new figure
plt.figure(figsize=(12,4))

# Display the spectrogram on a mel scale
plt.subplot(2,1,1)

# sample rate and hop length parameters are used to render the time axis
librosa.display.specshow(M, sr=sr, x_axis='time', y_axis='mel')

# plt.subplot(2,1,2)
# librosa.display.specshow(F, sr=sr, x_axis='time', y_axis='mel')
# # Put a descriptive title on the plot
plt.title('CQT')

# draw a color bar
plt.colorbar(format='%+02.0f dB')

# Make the figure layout compact
plt.tight_layout()
```

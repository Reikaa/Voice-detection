```python
# We'll need numpy for some mathematical operations
import numpy as np
import json

# Librosa for audio
import librosa

# matplotlib for displaying the output
import matplotlib.pyplot as plt
%matplotlib inline
# And seaborn to make it look nice
import seaborn
seaborn.set(style='ticks')

# and IPython.display for audio output
import IPython.display
```

```python
music = '3artists_output.wav'
```

```python
input1 = []
seconds_to_get = 1221
seconds_for_music = 2010

for i in range(seconds_for_music):
    y, sr = librosa.load('3artists_output.wav', duration=1.0, offset=i)
    input1.append(y)
    
print len(input1)
print len(input1[0])



```

```python
badoo = input1[400] # random 1sec of male sample
yance = input1[1300] # random 1sec of female sample
chetb = input1[1970]


erykah = librosa.feature.melspectrogram(y=badoo, sr=22050, S=None, n_fft=1024)
beyonce = librosa.feature.melspectrogram(y=yance, sr=22050, S=None, n_fft=1024)
chet = librosa.feature.melspectrogram(y=chetb, sr=22050, S=None, n_fft=1024)

```

```python
# Make a new figure
plt.figure(figsize=(15,10))

plt.subplot(6,1,1)
librosa.display.specshow(erykah, sr=sr, x_axis='time', y_axis='mel')
plt.subplot(6,1,2)
librosa.display.specshow(beyonce, sr=sr, x_axis='time', y_axis='mel')
plt.subplot(6,1,3)
librosa.display.specshow(chet, sr=sr, x_axis='time', y_axis='mel')

# draw a color bar
# plt.colorbar(format='%+02.0f dB')

# Make the figure layout compact
plt.tight_layout()
```
